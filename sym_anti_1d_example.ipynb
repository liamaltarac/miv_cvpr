```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'ieee'])
plt.rcParams.update({'figure.dpi': '100'})

def even_odd_decomposition(f):
    f = np.asarray(f)
    f_rev = f[::-1]
    even = (f + f_rev) / 2
    odd = (f - f_rev) / 2
    return even, odd

# Example
f =    np.array([0.18,0.43,0.13,0.81,0.37,0.76,0.94,0.79,0.42,0.26,0.60 ]) #np.array(np.random.random(11))
even, odd = even_odd_decomposition(f)

N = len(f)
# ---- Centered x-axis: -N/2 ... +N/2 ----
x = np.arange(-(N // 2), (N // 2) + 1)

plt.figure(figsize=(5, 4))

plt.plot(x, even, 's--', label='Even (symmetric)', color='red')
plt.plot(x, odd, 'd--', label='Odd (antisymmetric)', color='blue')

plt.xlabel('n', fontsize=14)
plt.ylabel('value', fontsize=14)

plt.ylim(bottom=min(f.min(), even.min(), odd.min(), 0), top=1.0)
plt.title('Symmetricâ€“Antisymmetric Decomposition of a Discrete Function')
plt.legend()
plt.grid(True, linestyle=':')

plt.axhline(y=0, color='black', linewidth=1.5, linestyle='--')
plt.tight_layout()

plt.savefig("sym_antisym_1d.pdf", bbox_inches='tight', dpi=600)
plt.show()
```
