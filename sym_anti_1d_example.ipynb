```python
import numpy as np



import matplotlib
import matplotlib.pyplot as plt
import scienceplots



plt.style.use(['science', 'ieee'])
plt.rcParams.update({'figure.dpi': '100'})



def even_odd_decomposition(f):
    """
    Decompose a discrete signal f into its even (symmetric) and odd (antisymmetric) parts.
    Works for 1D NumPy arrays.

    Even part:  (f[n] + f[-n]) / 2
    Odd part:   (f[n] - f[-n]) / 2
    """
    f = np.asarray(f)
    N = len(f)
    # Handle discrete symmetry around the center
    # Center the indices around 0
    f_rev = f[::-1]
    even = (f + f_rev) / 2
    odd = (f - f_rev) / 2
    return even, odd


# Example usage
f = np.array(np.random.random(11))
even, odd = even_odd_decomposition(f)
x = np.linspace(0, 1, len(f))
plt.figure(figsize=(5, 4))
#plt.plot(x, f, 'o-', label='Original f')
plt.plot(x, even, 's--', label='Even (symmetric)', color='red')
plt.plot(x, odd, 'd--', label='Odd (antisymmetric)' , color='blue')

plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

plt.ylim(bottom=min(f.min(), even.min(), odd.min(), 0), top=1.0)
plt.title('Symmetric-Antisymmetric Decomposition of a Discrete Function')
plt.legend()
plt.grid(True, linestyle=':')


plt.axhline(y=0, color='black', linewidth=1.5, linestyle='--')
plt.tight_layout()
plt.show()
plt.savefig(f"sym_antisym_1d.pdf", bbox_inches='tight', dpi=600)
```
